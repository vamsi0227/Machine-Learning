```python
# Lets try to analyze the values present in Attrition column of the employee dataset

# here we are using value_counts() which return the values related to each of the categories present in the attribute
x = data['Company'].value_counts()

# printing the values
print(x)
```

```python
plt.rcParams['figure.figsize'] = (4, 3)
x.plot(kind = 'bar', color ='green')

plt.title('Univariate Analysis on Attrition Column of Employee Dataset', fontsize = 20)
plt.xlabel('Company')
plt.ylabel('Values')
plt.show()
```

```python
# analyzing the attribute Department
plt.rcParams['figure.figsize']=(15, 7)
data['Company'].value_counts().plot(kind = 'bar', color = 'gold')
plt.ylabel('Values')
plt.xlabel('Different Companies')
#plt.xticks(rotation = 360)
plt.title('No. of Products in each Departments', fontsize =12)
plt.show()
```

```python
x = data['TypeName'].value_counts()

# printing the values
print(x)
```
